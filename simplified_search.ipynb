```python

def gaa(hoca:str):
    pubs = ""
    titles = ""
    for p in db.search(Query().author == hoca)[0].get("publications"):
        try:
            pubs+=p.get("abstract")
        except TypeError:
            pass
        try:
            titles+=p.get("title")
        except TypeError:
            pass
    return (pubs.lower(),titles.lower())
```

```python
def findall(k,t):
    if not isinstance(k,list):
        k = [k]
    return [re.findall(i,t) for i in k]
```

```python
def some_pooling_method(keyword:str):
    with open("topics.txt","r") as t:
        topics_str = t.read()
    topics = [i.split(",") for i in topics_str.split("\n")]
    for i in topics:
        if keyword in i:
            return i
    return [keyword]
```

```python
def new_search(keyword:str):
    rl = []
    pool = some_pooling_method(keyword.lower())
    for author in db:
        p,t=gaa(author.get("author"))
        result = findall(pool,p+t*5)
        rl.append([author.get("author"),result])
    score_list = [[i[0],sum([len(x) for x in i[1]])] for i in rl]
    score_dict = {}
    for i in score_list:
        score_dict[i[0]] = i[1]
    return score_dict
```

```python
rl = new_search("neural network")
```
